```python
###### looking at binary image of the actual video trying to see the rope segment and trying to isolate it: 

### chatgpt generated version of the code
'''looking at the difference of frames and seeing how they work with the use of adaptive thresholding using 
Gaussian and mean thresholding 
'''



import cv2

# Open the video file
cap = cv2.VideoCapture(path)

# Capture the first frame of the video
ret, frame1 = cap.read()
gray1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

while True:
    # Read the next frame of the video
    ret, frame2 = cap.read()
    if not ret:
        break
    
    # Convert the current frame to grayscale
    gray2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    
    # Calculate the absolute difference between the frames
    diff = cv2.absdiff(gray1, gray2)
    
    # Apply a threshold to filter out small changes
    thresh1 = cv2.adaptiveThreshold(diff,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
    
    thresh2 = cv2.adaptiveThreshold(diff,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
                                   cv2.THRESH_BINARY,11,2)
    
    
    # Set the current frame as the reference frame for the next iteration
    gray1 = gray2
    
    # Display or save the resulting difference image
    cv2.imshow('Difference Image Mean Thresh', thresh1)
    cv2.imshow('Difference Image Gaussian Thresh', thresh2)
    cv2.imshow("Actual Frame",frame2)
    
    if cv2.waitKey(100) & 0xFF == ord('q'):
        break

# Release the video capture and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
### this set of code is used to remove contours which are not stationary - Tken from OPenAI

import cv2

# define the ROI coordinates
x, y, w, h = 100, 200, 30, 30

# create a VideoCapture object
cap = cv2.VideoCapture(path)

# initialize the background subtractor object
backSub = cv2.createBackgroundSubtractorMOG2(history=10, varThreshold=50, detectShadows=False)

# initialize the Canny edge detection parameters
canny_min = 50
canny_max = 100

# initialize a list to store the previous contours
prev_contours = []

while True:
    # read the frame from the video
    ret, frame = cap.read()

    # check if the frame was successfully read
    if not ret:
        break

    # apply the background subtraction to obtain a foreground mask
    fgMask = backSub.apply(frame)

    # apply the mask to the ROI
    roi = fgMask[y:y+h, x:x+w]
    roi1 = frame[y:y+h, x:x+w]
    

    # apply Canny edge detection to the ROI
    edges = cv2.Canny(roi, canny_min, canny_max)

    # apply morphological operations to clean up the edges
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

    # find the contours in the edges image
    contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # initialize a list to store the current contours
    current_contours = []

    # loop over the contours
    for contour in contours:
        # compute the bounding box of the contour
        x, y, w, h = cv2.boundingRect(contour)

        # check if the contour is within the ROI
        if x >= 0 and y >= 0 and x+w <= roi.shape[1] and y+h <= roi.shape[0]:
            # add the contour to the current contours list
            current_contours.append(contour)

    # initialize a list to store the filtered contours
    filtered_contours = []

    # loop over the current contours
    for contour in current_contours:
        # check if the contour was not present in the previous frames
        if not any([cv2.matchShapes(contour, prev_contour, cv2.CONTOURS_MATCH_I2, 0.0) == 0 for prev_contour in prev_contours]):
            # add the contour to the filtered contours list
            filtered_contours.append(contour)

    # draw the filtered contours on the original frame
    frame_copy = frame.copy()
    cv2.drawContours(frame_copy[y:y+h, x:x+w], filtered_contours, -1, (0, 255, 0), 2)

    # update the previous contours list
    prev_contours = current_contours

    # display the frame with contours
    cv2.imshow('Frame', frame_copy)
    cv2.imshow("mask",fgMask)
    cv2.imshow("roi",roi1)
    

    # wait for the user to press a key
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

# release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
